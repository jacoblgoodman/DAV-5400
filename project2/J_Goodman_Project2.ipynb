### Data Dictionary:

Once you run the above (_don't forget to insert an API key_) we will now have a dataset with all rows of all orders between January 1st and march 30th 2018.  

​	This is 542,643 rows of data with 15 attributes which are as follows:

1. __invoice_line_no__: Concatenated invoice and line number associated with the liquor order. This provides a unique identifier for the individual liquor products included in the store order. 

   ​	data type = categorical 

2.  **date**: the date and time of the of the order

   ​	data type = time series /continuous 

3. **name**: Name of the store who ordered the liquor

   ​	data type = Categorical 

4. **city**: the city were the store who ordered the liquor is located

   ​	data type = Categorical 

5. **county**: the country where the store who ordered the liquor is located**

   ​	data type = Categorical

6. **zipcode**: zip code where the store who ordered the liquor is located

   ​	data type = Categorical 

7. **itemno**: The SKU (stock keeping unit) or item number for the individual liquor product ordered

   ​	data type = categorical 

8. **im_desc**: Description of the individual liquor product ordered

   ​	data type = Categorical

9. **vendor_name**: The vendor name of the company for the brand of liquor ordered

   ​	data type = categorical 

10. **category_name**: The Category of the liquor ordered

    ​	data type = Categorical 

11. **bottle_volume_ml** :  the Volume of each bottle order in milliliters 

    ​	data type = discrete

12. **pack**: the number of bottles in a case for the liquor ordered 

    ​	data type = Ordinal (it has 14 distinct values)

13. **sale_bottles**: The number of bottles of liquor ordered by the store

    ​	data type = discrete

14. **state_bottle_cost**: The amount that Alcoholic Beverages Division paid for each bottle of liquor ordered (cost)

    ​	data type = continuous 

15. **state_bottle_retail**: The amount the store paid for each bottle of liquor ordered (revenue)

    ​	data type = continuous 
16. **sale_dollars**: number of bottles X sale_bottles

    ​	data type = continuous

**Process**

To answer our research questions we will need to follow several steps:

**Preparation:**

1. A profit column will be created
2. A total profit per line will be created by `multiplying profit x bottle count`     

**Analysis:**

1. The relationship between bottle volume and profit will be plotted.
2. A group-by will be created and summary statistics created to evaluate;
   1. most profitable category
   2. the distribution std deviation of profitability

````python
#p1. create profit collumn
df['profit'] = df.state_bottle_retail-df.state_bottle_cost 

#p2. total profit collumn 
df['total_profit'] = df.profit*df.sale_bottles

#create profit precent by bottle
df['percent_profit_bottle']= df.profit/df.state_bottle_cost


#sanity check
display(df[['profit','total_profit','percent_profit']].describe())
````

````python
# caculated with boxplots

#sns boxplot
plt.figure(figsize=(26,12));
#plt.subplot(2,1,1);
sns.boxenplot('bottle_volume_ml','profit', data=df);
plt.ylim(top=250);
plt.title('seaborn');
````

````python
#plot with matplotlib 
plot_data = []
for v in np.nditer(df.bottle_volume_ml.unique()):
    d =  df.profit[df.bottle_volume_ml==v]
    plot_data.append(d)
plt.figure(figsize = (26,12));
ax = plt.boxplot(plot_data, labels=df.bottle_volume_ml.unique())
plt.ylim(top=300);
plt.ylabel('profit');
plt.xlabel('volume_ml');
plt.title('matplotlib');
````
