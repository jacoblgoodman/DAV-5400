## Research Questions

### How does weather affect ridership of the Citi bike system? 

There are several reasons why we would want to know how much of an effect weather has on ridership depending on the entity asking the question and the type of weather in question. We will focus on two different reasons for this analysis. 

#### Seasonal bikeshare (Citibike motivate)  :

#### Weather events competitor program:

### Historical weather data: 

Two sources of historical weather data will be pursued:

1. Scraping from [wunderground](https://www.wunderground.com/)

   Data from the Newark Liberty International weather station appears to be accessible for web scraping:

   Data conforms to the following syntax: https://www.wunderground.com/history/daily/us/ny/new-york-city/KEWR/date/{date: yyyy-m-d}

   Hourly readings are present in a html table id = 'history-observation-table'

2. API via [darksky](https://darksky.net/dev/docs/sources)

   Dark sky provides a comprehensive and cost effective ($0.0001 per call) API  which returns both observed and forecast values for every hour of requested day.

## Approach

For this project I will download Citibike data as .csv (comma separated values) files this data will be in several files so I will upload them into a data base to bind them together. I will then acquire weather data for the same time period as my Citbike data and store that in a .csv file

Once I have my data I will need to aggregate the data to be at the same grain (day) and perform EDA (exploratory data analysis) to visualize and verify the data, make any necessary modifications (such as replacing null values) and tidy before the final analysis.

For the final analysis, I will generate some summary information to help explain the data, ridership statistics by day of the week and by month in particular. I will then create visualizations to see how ridership is related to two types of weather: temperature and weather events (ie rain, snow, etc).

Finally, with temperature, and weather events as predictor variables we will see how well they can be utilized to predict daily ridership through a  regression model either poisson or negative bionmial depending on the distribution of the data.
