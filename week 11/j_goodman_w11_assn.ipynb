```python
#api to capture all best selers 
base_url = "https://api.nytimes.com/svc/books/v3/lists/current/"
# for testing with less time
#example="hardcover-fiction"

#final data list initialized 
data = []
# iteration over lists 
for index, item in df.list_name_encoded.iteritems():
    #sanity check print every list url 
    print(f'{base_url}{item}.json')
    #make request
    rp = requests.get(f'{base_url}{item}.json',params=querystring)
    #extend method to create single list 
    data.extend(rp.json()['results']['books'])
    #delay betweeen itersation as recomened by NY times documentation to prevent erro 429(to many requests)
    time.sleep(6)
#sanity check
len(data)
```
